```python
import pandas as pd
```

```python
titanic = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/titanic.csv", usecols = ["survived", "pclass", "sex", "age", "fare"])
```

```python
titanic.head(20)
```

```python
summary = titanic.groupby(["sex", "pclass"]).mean()
```

```python
summary
```

```python
summary.index #Notice that we have a multi-index
```

```python
#Slicing
summary.loc[("female", 2), :] #Select pclass two of sex group female
```

```python
#Slicing
summary.loc[("female", 2), "age"] #Select only age column on row pclass two of sex group female
```

```python
#Swap index
summary.swaplevel()
```

```python
#Swap index
summary.swaplevel().sort_index()
```
