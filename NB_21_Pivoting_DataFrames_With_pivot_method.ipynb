# Limitations of Pivot() Method and how to overcome them

```python
table2 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/table2.csv")
```

```python
table2.head(20) #Notice the duplicates on ALG and Bronze
```

```python
table2.tail(10)#Notice the duplicates in cCountry and Year
```

```python
table2.info()
```

```python
table2.shape
```

```python
#This produces an error #Index contains duplicate entries, cannot reshape
table2.pivot(index="Country", columns="Medal", values="Count")
```

# How to Solve the problem

## Use groupby

```python
table2.groupby(["Country", "Year", "Medal"]).Count.sum()
```

### Then unstack()

```python
table2.groupby(["Country", "Year", "Medal"]).Count.sum().unstack().fillna(0)
```

```python
table2_pivoted = table2.groupby(["Country", "Year", "Medal"]).Count.sum().unstack().fillna(0)
```
