# Outer Joins (without intersections) using merge() method

## All Athletes that won in only one edition but not both i.e left_only and right_only

### First perform an outer join

```python
men0408_outer = men2004.merge(men2008, how = "outer", on = "Athlete", suffixes=("_2004", "_2008"), indicator=True)
```

```python
men0408_outer
```

#### Then we exclude both

```python
men0408_outer._merge
```

```python
men0408_outer._merge != "both"
```

### Now using the loc operator

```python
men0408_outer.loc[men0408_outer._merge != "both"]
```

```python
men0408_outer.loc[men0408_outer._merge != "both"]
```

```python
men_outer = men0408_outer.loc[men0408_outer._merge != "both"]
```

# Left Joins (without intersections) using merge() method

## All Athletes that won in only men2004 edition i.e left_only 

### Again we perform an outer join first

```python
men04_left_only = men2004.merge(men2008, how = "outer", on = "Athlete", suffixes=("_2004", "_2008"), indicator=True)
```

```python
men04_left_only
```

```python
men04_left_only._merge == "left_only"
```

```python
men04_left_only.loc[men04_left_only._merge == "left_only"] #43 Athletes
```

```python
men04_left_only.loc[men04_left_only._merge == "left_only"]._merge.shape
```

# Joining on different Column labels and Indexes

## 1: Joining on different Column labels

```python
men2004.head()
```

```python
men2008.head()
```

```python
#Let change the columns of one dataset
men2004.columns = ["Names", "Medals"]
```

```python
men2004.head()
```

```python
men0408_join_diff_cols = men2004.merge(men2008, how = "outer", left_on = "Names", right_on = "Athlete", suffixes=("_2004", "_2008"), indicator=True)
```

```python
men0408_join_diff_cols
```

```python
#Let join the columns Names and Athlete along Names column
men0408_join_diff_cols.Names.fillna(men0408_join_diff_cols.Athlete, inplace=True)
```

```python
men0408_join_diff_cols
```

```python
#Let now drop the column Athlete
men0408_join_diff_cols.drop(["Athlete"], axis = 1, inplace=True)
```

```python
men0408_join_diff_cols
```
