# More on the PeriodIndex Object

## Pitfalls When Using the PeriodIndex Object

```python
temp2.head()
```

```python
temp2.info()
```

Assumming we want to aggregate the data with the resample method by monthly

#### Let see the behaviour when using Datetimeindex

```python
temp_m2 = temp2.resample("M", kind="timestamp").mean()
```

```python
temp_m2.head()
```

```python
temp_m2.info()
```

```python
temp_m2.index
```

```python
#Slicing Jan 2013
temp_m2.loc["2013-01"]
```

```python
#Slicing from May 2013 to August 2013
temp_m2.loc["2013-05" : "2013-08"]
```

```python
#Slicing the year 2013
temp_m2.loc["2013"]
```

#### Now - Let see the behaviour when using periodindex

```python
temp_m3 = temp2.resample("M", kind="period").mean()
```

```python
temp_m3.head() #Notice the day is missing - We now have just the Months
```

```python
temp_m3.info() #We now have PeriodIndex
```

```python
temp_m3.index
```

```python
#Slicing Jan 2013 - Notice with DatetimeIndex we get a DataFrame but with the PeriodIndex we get a Series
temp_m3.loc["2013-01"]
```

```python
#Slicing from May 2013 to August 2013 - Same Data Structure here
temp_m3.loc["2013-05" : "2013-08"]
```

#### Now Comes the Suprise

Notice we only get the data for Jan 2013 instead of the data for 12 months in the year 2013

```python
#Slicing the year 2013
temp_m3.loc["2013"]
```

### Solution

#### Changing the data back to timestamp using the to_timestamp() method

```python
temp_m4 = temp_m3.to_timestamp(how="start")
```

```python
temp_m4.head()
```

```python
temp_m4.info()
```

```python
temp_m4.index
```

```python
birthd2 = pd.date_range(end="2018-07-04", periods=10, freq=pd.DateOffset(years=1))
```

```python
birthd2
```

```python
temp_d.loc[birthd2] #KeyError: 'Passing list-likes to .loc or [] with any missing labels is no longer supported,
```

## Solution - Use the reindex() method

```python
temp_d.reindex(birthd2)
```

```python
#Using the method parameter to fill the NAN
temp_d.reindex(birthd2, method="bfill") #backfill / bfill: Use next valid observation to fill gap.
```

```python
#Notice the very first observation used above
temp_d.head()
```

```python
#Using the method parameter to fill the NAN
temp_d.reindex(birthd2, method="ffill") #pad / ffill: Propagate last valid observation forward to next valid.
```

```python
#Notice the very last observation used above
temp_d.tail()
```

```python
#Using the method parameter to fill the NAN
temp_d.reindex(birthd2, method="nearest") #nearest: Use nearest valid observations to fill gap.
```
