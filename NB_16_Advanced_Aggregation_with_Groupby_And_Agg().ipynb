```python
import pandas as pd
```

```python
titanic = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/titanic.csv", usecols=["survived", "pclass", "age", "sex", "fare"])
```

```python
titanic.head()
```

```python
titanic.groupby("sex").agg(["mean", "sum", "min", "max"])
```

# More flexible approach

```python
titanic.groupby("sex").agg({"survived" : ["sum", "mean"], "pclass" : "mean", "age" : ["mean", "median", "min", "max"], "fare" : ["mean", "min", "max"]})
```

# Groupby Aggregation with Re-Labelling - (Named Aggregation)

```python
titanic.groupby("sex").agg(survival_rate = ("survived", "mean"))#The label here is survival_rate
```

```python
titanic.groupby("sex").agg(survival_rate = ("survived", "mean"),
                          total_survivers = ("survived", "sum"),
                          mean_age_of_passengers = ("age", "mean"))
```
