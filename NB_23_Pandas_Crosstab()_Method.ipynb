```python
import pandas as pd
```

```python
titanic = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head(10)
```

```python
titanic.tail(10)
```

```python
titanic.info()
```

```python
#Here we pass the sex column as index and the pclass as as the columns - returns the frequencies of each group
pd.crosstab(titanic.sex, titanic.pclass)
```

### Similar with the groupby - unstack combo

```python
titanic.groupby(["sex", "pclass"]).pclass.count().unstack()
```

# Looking at the crosstab() other parameters

```python
pd.crosstab(titanic.sex, titanic.pclass, margins=True, margins_name="Totals")
```

```python
pd.crosstab(titanic.sex, titanic.pclass,margins=True, normalize=True).round(3) #returns the overal percentage
```

```python
#Returns the mean age  of each groups - sex and pclass #Notice how we pass the index and column explicitly and the age to the value parameter
pd.crosstab(index = titanic.sex, columns=titanic.pclass, values=titanic.age, aggfunc="mean").round(1)
```

### Alternative with the pivot_table() method

```python
titanic.pivot_table(index = "sex", columns="pclass", values="age", aggfunc="mean").round(1)
```

### Alternative with the groupby - unstack methods combo

```python
titanic.groupby(["sex", "pclass"]).age.mean().unstack().round(1)
```

## More advanced use of the crosstab() method

```python
summer = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/summer.csv")
```

```python
summer.head(10)
```

```python
summer.tail(10)
```

```python
summer.info()
```

```python
#Notice here we are passing summer.Year and summer.Country as multi-index
pd.crosstab(index = [summer.Year, summer.Country], columns=summer.Medal,
            values=summer.Athlete, aggfunc="count", margins=True).fillna(0)
```

### Alternative with the pivot_table() method

```python
summer.pivot_table(index = ["Year", "Country"], columns="Medal", values="Athlete",
                   aggfunc="count", margins=True, fill_value=0)
```

### Alternative with the groupby - unstack methods combo

```python
summer.groupby(["Year", "Country", "Medal"]).Medal.count().unstack(fill_value=0)
```

```python
test = summer.groupby(["Year", "Country", "Medal"]).Medal.count().unstack(fill_value=0)
```
