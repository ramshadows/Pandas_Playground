```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("Data1/Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.plot()
plt.show()
```

```python
titanic.plot(subplots = True, figsize=(15, 12), sharex = False)
plt.show()
```

```python
xticks = [x for x in range(0, 901, 50)]
yticks = [y for y in range(0, 81, 5)]
```

```python
titanic.age.plot(figsize=(16,12), fontsize = 20, color = "green", linestyle="-", xticks = xticks, yticks = yticks,
                rot = 45)
plt.title("Titanic - Age", fontsize=15)
plt.legend(loc = "upper right", fontsize=13 )
plt.xlabel("Passenger No", fontsize=13)
plt.ylabel("Age", fontsize=13)
plt.show()
```

# Histograms

```python
titanic.age.plot(kind="hist", fontsize=15, figsize=(15, 12))
plt.show()
```

```python
titanic.age.value_counts(bins = 80)
```

```python
titanic.age.plot(kind="hist", fontsize=15, figsize=(15, 12), bins = 80)
plt.show()
```

# Method 2 : Plotting Histograms

```python
titanic.age.hist(figsize=(15,12), bins = 80, xlabelsize = 15, ylabelsize= 15)
plt.show()
```

# Method 3

```python
plt.figure(figsize=(15,12))
plt.hist(titanic.age.dropna(), bins=80, density = False, cumulative = False)
plt.show()
```

# Scatterplots

```python
titanic.head()
```

```python
titanic.plot(kind="scatter", figsize=(15,8), fontsize=15, x = "age", y = "fare", color = "green", marker = "D", s = 5)
plt.show()
```

```python
#Adding 3D by adding another column on the c attribute
titanic.plot(kind="scatter", figsize=(15,8), fontsize=15, x = "age", y = "fare", c = "survived", marker = "D",
             s = 30, colormap = "viridis")
plt.show()
```

```python
#Adding 3D by adding another column on the c attribute
titanic.plot(kind="scatter", figsize=(15,8), fontsize=15, x = "age", y = "fare", c = "pclass", marker = "D",
             s = 30, colormap = "viridis")
plt.show()
```
