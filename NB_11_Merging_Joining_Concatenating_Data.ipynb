# Merging, Joining and Concatenating Data

```python
import pandas as pd

pd.options.display.max_rows = None
```

```python
men2004 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/men2004.csv")
```

```python
men2004
```

```python
men2008 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/men2008.csv")
```

```python
men2008
```

## Method 1: Using the append() method

```python
men2004.append(men2008, ignore_index=True)#Append rows of `other` to the end of caller, returning a new object. Columns in `other` that are not in the caller are added as new column
```

## Method 2: Using concat() method - Best Option

```python
pd.concat([men2004, men2008], ignore_index=False, keys=[2004, 2008])#Setting the ignore_index to False keeps the original indexes of the two data frames
```

```python
pd.concat([men2004, men2008], ignore_index=True, keys=None)
```

```python
men0408 = pd.concat([men2004, men2008], ignore_index=False, keys=[2004, 2008], names=["Year"])
```

```python
men0408
```

```python
men0408.info()
```

###### Now we can decide to play with our index to suit our preference

###### Example

```python
men0408.reset_index()
```

```python
men0408.reset_index().drop(columns="level_1")
```

# Arithmetic with Pandas data object

```python
topfive_2004 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/topfive_2004.csv", index_col = "Athlete")
```

```python
topfive_2004
```

```python
topfive_2008 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/topfive_2008.csv", index_col = "Athlete")
```

```python
topfive_2008
```

```python
#Inspecting the two dataset we see that topfive_2008 has column bronze in lower case
#Let re-align the dataset before doing the arithmetic for better results
topfive_2008.rename(columns={"bronze" : "Bronze"}, inplace=True)
```

```python
topfive_2008
```

## Let now do the addition of the two dataet with add() Method

```python
topfive_2004.add(topfive_2008, fill_value=0.0) #Fill NAN values with zero
```

```python
topfive_2004.sub(topfive_2008, fill_value=0)
```
