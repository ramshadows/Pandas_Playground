# Transformation or Mapping with map()

```python
import pandas as pd
```

```python
summer = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/summer.csv")
```

```python
summer.head()
```

```python
#Let pick a random sample using sample of 7 rows
sample = summer.sample(n=7, random_state=123).sort_values(by = "Year")
```

```python
sample
```

```python
#Dictionary of City and Country
city_country = {"Paris" : "France", "Mexico" : "Mexico", "Montreal" : "Canada", "Moscow" : "Russia", "Barcelona" : "Spain", "Athens" : "Greece"}
```

```python
city_country
```

```python
#Mapping our City to Country in our Sample dataset
sample.City.map(city_country)
```

```python
#Adding a new column for the cities
sample["Country"] = sample.City.map(city_country)
```

```python
sample
```

```python
#Using the titanic Dataset
titanic = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/titanic.csv")
```

```python
titanic.head()
```

```python
#Mapping pclass to their respective string 
mapper = {1 : "First", 2 : "Second", 3 : "Third"}
```

```python
mapper
```

```python
titanic.pclass.map(mapper)
```

```python
titanic.pclass = titanic.pclass.map(mapper)
```

```python
titanic.head()
```

# Conditional Transformation

```python
import numpy as np
```

```python
#Check if a passenger had a sibling or spouse
titanic.sibsp == 0
```

```python
np.where(titanic.sibsp == 0, "Yes", "No")
```

```python
pd.Series(np.where(titanic.sibsp == 0, "Yes", "No"))
```

```python
titanic["alone"] = pd.Series(np.where(titanic.sibsp == 0, "Yes", "No"))
```

```python
titanic.head(10)
```

```python
titanic["child"] = pd.Series(np.where(titanic.age < 18, "Yes", "No"))
```

```python
titanic.head(10)
```

```python
titanic["adult"] = pd.Series(np.where(titanic.age > 18, "Yes", "No"))
```
