```python
plt.figure(figsize = (10,6))
plt.hist(y, bins=100, label="data") #Histogram
plt.title("Frequency Distribution of y")
plt.xlabel("y")
plt.ylabel("Frequencies")
plt.legend()
plt.show()
```

```python
#Adding a vertical line into our graph
plt.figure(figsize = (10,6))
plt.hist(y, bins=100, label="data") #Histogram
plt.title("Frequency Distribution of y")
plt.vlines(np.mean(y), 0, 350, label="Mean", linestyles="dotted", colors="black")#Vertical line
plt.xlabel("y")
plt.ylabel("Frequencies")
plt.legend()
plt.show()
```

```python
#Create 20 normally distributed numbers with mean 10 and std 2 and shape of(2,10)
m = np.random.normal(10, 2, 20).reshape(2,10)
```

```python
m
```

```python
#Slicing first row of m - nB contains 10 elements
a = m[0]
```

```python
a
```

```python
#Slicing second row of m
b = m[1]
```

```python
b
```

```python
plt.figure(figsize = (10,6))
plt.scatter(a, b)
plt.show()
```

```python
#Linear Regression - Fits a straight line into our scatter plots
#Note: 1=lienear, 2=Quadratic, 3=Cubic
#Note b=(12.5 - 0.15)*a
reg1 = np.polyfit(a,b, 1)
```

```python
reg1
```

```python
#Creating x and y for regression line
x = np.linspace(min(a), max(a), num=100)
y = np.polyval(reg1, x)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(a, b, label="Data") #Data points
plt.plot(x, y, 'b--', label="linear") #Linear Regression - Best line that fits to the data
plt.legend()
plt.show()
```

```python
#Getting quadratic by setting degree to 2
reg2 = np.polyfit(a,b, 2)
```

```python
reg2 #Function: b = (16.9856 + 0.0535 * a**2 - 1.1738 * a)
```

```python
#Plotting our quadratic expression
plt.figure(figsize = (10,6))
plt.scatter(a, b, label="Data") #Data points
plt.plot(x, np.polyval(reg2, x), 'm-.', label="quadratic") #Quadratic expression
plt.legend()
plt.show()
```

```python
#Getting a cubic by setting degree to 3
reg3 = np.polyfit(a,b, 3)
```

```python
reg3 #Function: b = 92.255 - 0.1178 * a**3 + 3.2678 * a **2 - 28.8996 * a
```

```python
#Plotting our cubic expression
plt.figure(figsize = (10,6))
plt.scatter(a, b, label="Data") #Data points
plt.plot(x, np.polyval(reg3, x), 'g.', label="cubic") #Cubic expression
plt.legend()
plt.show()
```

```python
#Combining our curves
plt.figure(figsize = (10,6))
plt.scatter(a, b, label="Data") #Data points
plt.plot(x, y, 'b--', label="linear") #Linear Regression - Best line that fits to the data
plt.plot(x, np.polyval(reg2, x), 'm-.', label="quadratic") #Quadratic expression
plt.plot(x, np.polyval(reg3, x), 'g.', label="cubic") #Cubic expression
plt.legend()
plt.show()
```

```python
#Trying a perfect regression
reg_perfect = np.polyfit(a,b, len(a)-1) #polynomial of degree 9
```

```python
reg_perfect
```

```python
#Plotting
plt.figure(figsize = (10,6))
plt.scatter(a, b, label="Data") #Data points
plt.plot(x, np.polyval(reg_perfect, x), 'b--', label="perfect")
plt.legend()
plt.show()
```

```python
#Zooming with ylim
#Plotting
plt.figure(figsize = (10,6))
plt.scatter(a, b, label="Data") #Data points
plt.plot(x, np.polyval(reg_perfect, x), 'b--', label="perfect")
plt.ylim(min(b)-0.5, max(b)+0.5) #Zooms
plt.legend()
plt.show()
```
