# The shift() method on Time Series

### Calculate the Absolute/Relative Change/Difference

```python
#Copy the AAPL Stock and Transform the Series to a DataFrame
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
#Notice the first value is NAN bcoz we have no value behond it to shift with
aapl.shift(periods=1) #Shift index by desired number of periods =>1 with an optional time `freq`.
```

```python
aapl.shift(periods=2)
```

```python
aapl.shift(periods=-1)
```

```python
aapl["lag1"] = aapl.shift(periods=1)
```

```python
aapl.head()
```

```python
#Calculating the Daily increase/Decrease the Difference
aapl.AAPL.sub(aapl.lag1)
```

```python
aapl["diff"] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
#Calculating the Percetage Change or Return
aapl["percentage_change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

# The methods diff() and pct_change()

```python
aapl["diff2"] = aapl.AAPL.diff(periods=1)
```

```python
aapl.head()
```

```python
aapl["percetage_change2"] = aapl.AAPL.pct_change(periods=1).mul(100)
```

```python
aapl.head()
```

### Calculating Monthly percetage change

```python
#First resample the AAPL column
aapl.AAPL.resample("BM") # BM => Business days for the month
```

```python
#We are interested on the very last stock of the month
aapl.AAPL.resample("BM").last()
```

```python
#Now calculating the return
aapl.AAPL.resample("BM").last().pct_change(periods = 1).mul(100)
```

```python
#Standard Deviation of annual Return - Square root of Variance
std_annual_return = np.sqrt(var_annual_return)
std_annual_return
```

```python
daily_returns.describe()
```

```python
#Transposing to have stocks as our index
daily_returns.describe().T
```

```python
#Slicing only the mean and std columns
daily_returns.describe().T.loc[:, ["mean", "std"]]
```

```python
summary_stats = daily_returns.describe().T.loc[:, ["mean", "std"]]
```

```python
summary_stats.head()
```

```python
#Annualizing the summary Stats for mean and std
summary_stats["mean"] = summary_stats["mean"] * 252
summary_stats["std"] = summary_stats["std"] * np.sqrt(252)
```

```python
summary_stats
```

```python
summary_stats.plot(kind="scatter", x="std", y="mean", figsize=(12,8), s=50, fontsize=15)
plt.xlabel("Annual. Risk(std)", fontsize=15)
plt.ylabel("Annual. Return", fontsize=15)
plt.title("Risk Vs Return", fontsize=20)
plt.show()
```

```python
#Adding labels to the scatters using the annotate() method -Annotate the point *xy* with text *text*.
summary_stats.plot(kind="scatter", x="std", y="mean", figsize=(12,8), s=50, fontsize=15)
for i in summary_stats.index:
    plt.annotate(i, xy=(summary_stats.loc[i, "std"]+0.002, summary_stats.loc[i, "mean"]+0.002), size=15)
plt.xlabel("Annual. Risk(std)", fontsize=15)
plt.ylabel("Annual. Return", fontsize=15)
plt.title("Risk Vs Return", fontsize=20)
plt.show()
```

```python
close.head(20)
```

```python
#Using back fill method - Takes the next value to the last NA value which is not NA and back fills the NA Values
close.fillna(method='bfill')
```

```python
#Using forward fill method - Takes the first value next to the first occurence of NA value which is not NA and forward fills the NA Values
close.fillna(method='ffill')
```
