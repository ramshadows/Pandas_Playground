# Count Plot

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic, x="sex")
plt.show()
```

```python
#Splitting our plots more with hue parameter
plt.figure(figsize=(12,8))
sns.countplot(data=titanic, x="sex", hue="pclass")
plt.show()
```

```python
#Customizing with sns.set()
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.countplot(data=titanic, x="sex", hue="pclass")
plt.show()
```

# Categorical Plots

## 1: Strip Plot

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.stripplot(data=titanic, x="sex", y="age", jitter=False,  hue=None, dodge=False)
plt.show()
```

```python
#Setting jitter to True - 
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.stripplot(data=titanic, x="sex", y="age", jitter=True,  hue=None, dodge=False)
plt.show()
```

```python
#Setting dogde to True - Separates the hue plotting
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.stripplot(data=titanic, x="sex", y="age", jitter=True,  hue="pclass", dodge=True)
plt.show()
```

# 2: Swarm plot

```python
#Note almost similar to strip plot plus has no jitter parameter
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.swarmplot(data=titanic, x="sex", y="age",  hue="pclass", dodge=True)
plt.show()
```

# 3: Violin Plot - Shows the frequencies distribution

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.violinplot(data=titanic, x="sex", y="age", jitter=True,  hue="pclass", dodge=True)
plt.show()
```

# 4: Combining the Swarm Plot and Violin plot

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.violinplot(data=titanic, x="sex", y="age", jitter=True,  hue="pclass", dodge=True)
sns.swarmplot(data=titanic, x="sex", y="age",  hue="pclass", dodge=True, color="black")
plt.show()
```

### Swapping x and y axis values

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.violinplot(data=titanic, x="pclass", y="age", jitter=True,  hue="sex", dodge=True, split=False)
plt.show()
```

```python
#Setting split to True
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.violinplot(data=titanic, x="pclass", y="age", jitter=True,  hue="sex", dodge=True, split=True)
plt.show()
```

# 5: Bar Plots

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.barplot(data=titanic, x="pclass", y="age", hue="sex", dodge=True)
plt.show()
```

# 6: Point Plot

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.pointplot(data=titanic, x="pclass", y="age", hue="sex", dodge=True)
plt.show()
```

```python
sns.set(font_scale=2, palette="viridis")
sns.lmplot(data=titanic, x="age", y="survived", height=8, aspect=1, col=None, logistic=False)
plt.show()
```

```python
#Setting column parameter
sns.set(font_scale=2, palette="viridis")
sns.lmplot(data=titanic, x="age", y="survived", height=8, aspect=1, col="sex", logistic=False)
plt.show()
```

```python
#Setting column parameter and logistic regression to True
sns.set(font_scale=2, palette="viridis")
sns.lmplot(data=titanic, x="age", y="survived", height=8, aspect=1, col="sex", logistic=True)
plt.show()
```

```python
#Setting column parameter and logistic regression to True
#Splitting by pclass column
sns.set(font_scale=2, palette="viridis")
sns.lmplot(data=titanic, x="age", y="survived", height=8, aspect=1, col="pclass", logistic=True)
plt.show()
```

# Matrix Plots / Heatmaps

```python
#Splitting the sex and pclass with pd.crosstab
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot=False, fmt="d", cmap=None, vmax=None)
plt.show()
```

```python
#Setting annot to True
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot=True, fmt="d", cmap=None, vmax=None)
plt.show()
```

```python
#Setting color map - cmap to Reds
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot=True, fmt="d", cmap="Reds", vmax=None)
plt.show()
```

```python
pd.crosstab(titanic.sex, titanic.pclass, values=titanic.survived, aggfunc="mean")
```

```python
#Setting color map - cmap to Reds
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette="viridis")
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass, values=titanic.survived, aggfunc="mean"), annot=True, cmap="Reds", vmax=None)
plt.show()
```
