# Melting DataFrames with melt() method

```python
import pandas as pd
```

```python
table_2012 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/table_2012.csv")
```

```python
table_2012
```

```python
table_2012.shape
```

```python
table_2012.melt(id_vars="Country", value_vars=["Gold", "Silver", "Bronze"])#Notice how we are melting the columns Gold, Silver and Brnze to a single column variable
```

```python
#Customizing the Melted Columns names
table_2012.melt(id_vars="Country", value_vars=["Gold", "Silver", "Bronze"], var_name="Medal", value_name="Number")
```

```python
melted_2012 = table_2012.melt(id_vars="Country", value_vars=["Gold", "Silver", "Bronze"], var_name="Medal", value_name="Number")
```

```python
melted_2012.shape
```

```python
melted_2012
```

### Resetting Our DataFrame with pivot_table

```python
melted_2012.pivot_table(index="Country", columns="Medal", values="Number").sort_values("Gold", ascending=False)
```
