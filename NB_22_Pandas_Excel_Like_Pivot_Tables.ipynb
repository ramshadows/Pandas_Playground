```python
import pandas as pd
```

```python
table2 = pd.read_csv("Data1/Course_Material_Part2/Video_Lecture_NBs/table2.csv")
```

```python
table2.head(10) #Notice the duplicate entries in Country
```

```python
table2.pivot_table(index="Country", columns="Medal", values="Count", aggfunc="mean", fill_value=0).round(2)
```

```python
table2.pivot_table(index="Country", columns="Medal", values="Count", aggfunc="sum", fill_value=0).round(2)
```

# Alternative Method with a combination of groupby and unstack method

```python
table2.groupby(["Country", "Medal"]).Count.sum().unstack(fill_value=0)
```

# pivot_table() Method - Additional feature - margins and margins_name parameter

```python
table2.pivot_table(index="Country", columns="Medal", values="Count", aggfunc="sum", fill_value=0, margins=True, margins_name="Sub Totals").round(2)
```

# Similar implemetation with the groupby - unstack() combo

```python
agg = table2.groupby(["Country", "Medal"]).Count.sum().unstack(fill_value=0)
```

```python
agg.head()
```

```python
#Create a new column - "Sub Totals" as  sum of each row
agg["Sub Totals"] = agg.sum(axis=1)
```

```python
agg.head(10)
```

```python
#Create a new row - "Sub Totals" at the botton as sum of all columns
agg.loc["Sub Totals"] = agg.sum(axis=0)
```
