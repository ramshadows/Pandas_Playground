## Step 2

### Creating the Event_Gender Column

We need to determine for each row/medal, whether the medal was awarded in a Men, Women, or Mixed Event

```python
summer["Event_Gender"] = summer.Gender
```

```python
summer.head()
```

```python
summer.tail()
```

1: The Event column contains the String Mixed

```python
summer.Event.str.lower().str.contains("mixed").sum()
```

2: The Event column contains the String "pairs"

```python
summer.Event.str.lower().str.contains("pairs").sum()
```

3: All "Equestrians" have been mixed Events

```python
summer.Sport.str.lower().str.contains("equestrian").sum()
```

4: All Sailing Events before 1988 have been Mixed Events

```python
((summer.Sport.str.lower().str.contains("sailing")) & (summer.Year < 1988)).sum()
```

Creating Our Masks

```python
mask1 = summer.Event.str.lower().str.contains("mixed")
mask2 = summer.Event.str.lower().str.contains("pairs")
mask3 = summer.Sport.str.lower().str.contains("equestrian")
mask4 = ((summer.Sport.str.lower().str.contains("sailing")) & (summer.Year < 1988))
```

```python
#Spliting our summer datasets on rows where all the 4 masks conditions are met and overriding the Event_Gender #with String X to mean mixed event

summer.loc[mask1 | mask2 | mask3 | mask4, "Event_Gender"] = "X"
```

```python
#Inspecting our Changes
summer.Event_Gender.value_counts()
```

# Step 3

## Identifying all Unique Events and Count the number of medals in each event -New Column - Event_Medals

```python
summer["Event_Medals"] = summer.groupby(["Year", "Sport", "Discipline", "Event", "Event_Gender"]).Medal.transform("count")
```

```python
summer.head()
```

```python
summer.Event_Medals.value_counts().sort_index()
```

```python
summer.loc[summer.Event_Medals == 5]
```

# Step 4

## Identifying Team Events

Note: All medals or rows that were awarded in event with more than 5 medals shall be deemed Team Events

```python
summer["Team"] = pd.Series(np.where(summer.Event_Medals > 5, "Yes", "No"))
```

```python
summer.head()
```

```python
summer.Team.value_counts()
```

# Step 5

## Removing Duplicated Medals in Team_Events

The subset for determing duplicates shall be determined by columns - Year, Sport, Discipline, Country, Event,
Event_Gender

```python
summer.reset_index(inplace=True)
```

```python
summer.head()
```

```python
single_events = summer.loc[summer.Team == "No"].copy()
```

```python
single_events.shape
```

```python
team_events = summer.loc[summer.Team == "Yes"].copy()
```

```python
team_events.shape
```

```python
team_events.drop_duplicates(subset=["Year", "Sport", "Discipline", "Country", "Event", "Event_Gender", "Medal"], inplace=True)
```

```python
team_events.shape
```

```python
#Concate the team_events and single_events DFs
pd.concat([single_events, team_events]).shape
```

```python
summer_new = pd.concat([single_events, team_events])
```

```python
summer_new.set_index("index", inplace=True)
```
