# Ranking DataFrames with rank()

```python
sales = pd.Series([134, 567, 783, 112, 230], index=["Mon", "Tue", "Wed", "Thu", "Fri"], name = "Sales")
```

```python
sales
```

```python
sales.sort_values(ascending=True)
```

```python
sales.rank()
```

```python
sales.rank(ascending=False).sort_values()
```

```python
sales.rank(ascending=False, method="min").sort_values(ascending=True)
```

```python
sales.rank(ascending=False, method="min", pct=True).sort_values(ascending=True)
```

```python
titanic.fare.rank(ascending=False)
```

# User defined functions with apply() and applymap()

```python
sales = pd.read_csv("Data1/Video_Lecture_NBs/sales.csv", index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis = 0)
```

```python
sales.min(axis = 1)
```

```python
def range(series):
    return series.max() - series.min()
```

```python
sales.apply(range, axis=0)
```

```python
sales.apply(range, axis=1)
```

```python
sales.apply(lambda series: series.max() - series.min(), axis=0)
```

```python
titanic_1.loc[1]
```

```python
titanic_1.loc[[1,2]].sort_index()
```

```python
titanic_1.loc[:2]
```

```python
titanic_1.loc[(1, "female")]
```

```python
titanic_1.loc[(1, "female"), ["age"]]
```

```python
titanic_1.loc[(1, "female"), ["age", "fare"]]
```

```python
titanic_1.loc[([1,2], "female"), ["age", "fare"]].sort_index()
```

```python
titanic_1.loc[([1,2], "female"), :].sort_index()
```

```python
titanic_1.sort_index(inplace = True)
```

```python
titanic_1
```

```python
titanic_1
```

```python
titanic_1.loc[(slice(None), slice("female")), :] #Select all female passengers in all pclass
```

```python
titanic_1.loc[(slice(1), slice("female")), :] #Select all female passengers in  pclass 1
```

```python
titanic_1.loc[(slice(2,2), slice("female")), :] #Select all female passengers in  pclass 2
```

```python
titanic_1.loc[(slice(3,3), slice("female")), :] #Select all female passengers in  pclass 3
```

```python
titanic_1.loc[(slice(2,2), slice("male","male")), :] #Select all male passengers in  pclass 2
```

```python
titanic_1.loc[(slice(3,3), slice("male","male")), :] #Select all male passengers in  pclass 3
```

```python
titanic_1.loc[(slice(2,3), slice("male","male")), :] #Select all male passengers in  pclass 2 & 3
```

```python
titanic_1.loc[(slice(1,3), slice("male","male")), :] #Select all male passengers in  pclass 1 & 3
```
